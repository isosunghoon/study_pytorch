## Processing Data using Pytorch

Dataset, Dataloader class를 사용하여 데이터를 다룬다. Dataset은 데이터와 라벨을 묶어둔 것, Dataloader는 모델에 입력하기 편리하게 만들어둔 것이다.

#### Load Built-in Dataset

Mnist, CIFAR-10, Imagenet, COCO 같이 유명한 데이터셋은 아주 쉽게 불러올 수 있다

```python
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor


training_data = datasets.FashionMNIST(
    root="data", #데이터 위치 
    train=True,
    download=True, #데이터 없을 경우 다운로드 받는다
    transform=ToTensor(), #데이터에 할 transform
    target_transform = None #데이터 라벨에 할 transform
)

test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)
```

```python
import matplotlib.pyplot as plt

labels_map = {
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}

for i in range(1, 9):
    plt.subplot(2, 4, i)
    img, label = training_data[i]
    plt.title(labels_map[label])
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="gray")
plt.show()
```
